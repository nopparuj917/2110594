# Unigram

In this section, we will demonstrate how to build a unigram language model <br>
**Important note:** <br>
**\<s\>** = sentence start symbol <br>
**\</s\>** = sentence end symbol 

```python
def getUnigramModel(data):
    model = defaultdict(lambda: 0)
    word_count =0
    for sentence in data:
        sentence +=  u'|</s>' #for unigram model we can always ignore <s>, since p(w0=<s>)=1
        for w1 in sentence.split('|'):
            model[w1] +=1.0
            word_count+=1
    for w1 in model:
        model[w1] = model[w1]/(word_count)
    return model
```

```python
model = getUnigramModel(train)
```

```python
def getLnValue(x):
    if x >0.0:
        return math.log(x)
    else:
        return math.log(unk_value)
```

## TODO #3 **Create a Bigram Model**

```python
def getBigramModel(data):
    unigram_count = defaultdict(lambda: 0.0)
    bigram_count = defaultdict(lambda: 0.0)

    for sentence in data:
        for w1,w2 in bigrams(sentence.split("|"), pad_right=True, pad_left=True):
            unigram_count[w1] += 1
            bigram_count[w1, w2] += 1
    def model(w0, w1):
        # if unigram_count[w0] > 0 retrun bigram_count / unigram
        if unigram_count[w0] > 0 :
            return bigram_count[w0, w1] / unigram_count[w0]
        else:
        #else return unknown value
            return unk_value
    return model

model = getBigramModel(train)
```

## TODO #4 **Calculate Perplexity for Bigram Model**



```python
from math import exp
def calculate_sentence_ln_prob(sentence, model):
    ln_prob = 0
    #generate bigram and caculate ln_prob by using function "getLnValue"
    for w1,w2 in bigrams(sentence.split("|"), pad_right=True, pad_left=True):
        ln_prob += getLnValue(model(w1, w2))
        
    return ln_prob

def perplexity(test,model):
    ln_prob = 0
    word_count = 0
    #count word in sentence
    #and calculate ln_prob using function "calculate_sentense_ln_prob"
    for sentence in test:
        ln_prob += calculate_sentence_ln_prob(sentence, model)
        word_count += len(sentence.split("|"))
        word_count += 1
    #calculate perplexity by e^(-ln_prob / wordcount)
    return exp(-ln_prob / word_count)
    
```

# Smoothing

Usually any ngram models have a sparsity problem, which means it does not have every possible ngram of words in the dataset. Smoothing techniques can alleviate this problem. In this section, you will implement two basic smoothing methods laplace smoothing and interpolation for bigram.

## TODO #5 **Bigram with Laplace smoothing (Add-One Smoothing)**

```python
# Laplace Smoothing
def getBigramWithLaplaceSmoothing(data):
    #initializes unigram_count and bigram_count which return a defult value of 0.0 for any undefined keys.
    unigram_count = defaultdict(lambda: 0.0)
    bigram_count = defaultdict(lambda: 0.0)
    #extract all the bigarms and increments the count of each biagram_count and also unigram_count.
    for sentence in data:
        for w1, w2 in bigrams(sentence.split("|"), pad_right=True, pad_left=True):
            unigram_count[w1] += 1
            bigram_count[w1, w2] += 1
    # for every unigram add the unmber of it's count, effectively performing Laplace smoothing. 
    for i in unigram_count:
        unigram_count[i] += len(unigram_count)

    def model(w0, w1):
    # if count of w0 and w1 is greater than 0 it will return bigram count +1 / unigram
    #else return unkown value
        if unigram_count[w0] > 0 and bigram_count[w0, w1] > 0:
            return (bigram_count[w0, w1] + 1) / unigram_count[w0]
        return unk_value
    return model


model = getBigramWithLaplaceSmoothing(train)
print(perplexity(train, model))
print(perplexity(test, model))

# 974.8134581679766
# 1098.1622194979489
```

## TODO #6 **Bigram with Interpolation**
lambda value is 0.7 for bigram, 0.25 for unigram, and 0.05 for unknown word

```python
#interpolation
def getBigramWithInterpolation(data):
    #Fill code here
    #initializes unigram_count and bigram_count which return a defult value of 0.0 for any undefined keys.
    #and cnt is set to 0.
    unigram_count = defaultdict(lambda: 0.0)
    bigram_count = defaultdict(lambda: 0.0)
    cnt = 0
    
    #extract all the bigarms and increments the count of each biagram_count and also unigram_count.
    #increemtns cnt by 1 after processing a sentence cnt decrement by 1
    for sentence in data:
        for w1, w2 in bigrams(sentence.split("|"), pad_right=True, pad_left=True):
            unigram_count[w1] += 1
            bigram_count[w1, w2] += 1
            cnt += 1
        cnt -= 1
    
    #unigram count w0 and bigram count w0 w1 greater than 0 return
    #prob value follow this formula
    # 0.7bigram / unigram + 0.25unigram / cnt + 0.05unk value0.7bigram / unigram + 0.25unigram / cnt + 0.05unk value
    #else return unk value
    def model(w0, w1):
        if unigram_count[w0] > 0 and bigram_count[w0, w1] > 0:
            return (0.7 * bigram_count[w0, w1]  / unigram_count[w0]) + (0.25 * unigram_count[w1] / (cnt)) + (0.05 * unk_value)
        return unk_value
    return model

    
model = getBigramWithInterpolation(train)
print(perplexity(train,model))        
print(perplexity(test,model))

# 73.38409869825665
# 172.67485908813356
```

```python
def generate_seq(context, itos, max_new_token = 10):
  encoded = vocab(context)
  with torch.no_grad():
      for i in range(max_new_token):
          src = torch.LongTensor([encoded]).to(model.device)
          prediction = model(src)
          probs = torch.softmax(prediction[:, -1] / 1, dim=-1)  
          prediction = torch.multinomial(probs, num_samples=1).item()    
          
          while prediction == vocab['<unk>']:
              prediction = torch.multinomial(probs, num_samples=1).item()

          if prediction == vocab['<eos>']:
              break

          encoded.append(prediction)

  return "|".join([itos[e] for e in encoded])
```

```python
context = ["วัน", "จันทร์"]
generate_seq(context, itos, 20)
```
